```python
# câu 2a
def connect(dsn):
    conn = ibm_db_dbi.connect(dsn, "", "")
    return conn
```

```python
conn = connect(dsn)
stmt = '''
SELECT "Ma_sinh_vien", "Ho"||' '||"Ten" AS "Ho_ten", 
CASE 
	WHEN "Gioi_tinh" THEN 'Nam' 
	ELSE 'Nữ'
END AS "Gioi_tinh", 
VARCHAR_FORMAT("Ngay_sinh", 'DD/MM/YYYY HH24:MI:SS') AS "Ngay_sinh",
VARCHAR_FORMAT("Hoc_bong",'9,999,999.99') AS "Hoc_bong_"
FROM "SINH_VIEN"
ORDER BY "Hoc_bong"
'''
df = pd.read_sql(stmt, conn)
df.head()
```

```python
# câu 2b
conn = connect(dsn)
stmt = '''
SELECT *
  FROM "SINH_VIEN"
  WHERE "Hoc_bong">0
  ORDER BY "Hoc_bong" DESC
  FETCH FIRST 10 ROWS ONLY;
'''
df = pd.read_sql(stmt, conn)
df.head()
```

```python
# câu 3a
conn = connect(dsn)
stmt = '''
SELECT s."Ma_sinh_vien", s."Ho", s."Ten", s."Ma_khoa", k."Ten" AS "Ten_khoa"
  FROM "SINH_VIEN" s INNER JOIN "KHOA" k
  ON s."Ma_khoa"=k."Ma_khoa"
'''
df = pd.read_sql(stmt, conn)
df.head()
```

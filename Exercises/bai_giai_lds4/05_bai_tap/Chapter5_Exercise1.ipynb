```python
# Câu 2a: Liệt kê danh sách sinh viên gồm các cột Mã, Họ và Tên, Ngày sinh, Giới tính, Học bổng. Có sắp tăng theo cột học bổng.
def connect(dsn):
    conn = ibm_db_dbi.connect(dsn, "", "")
    return conn
```

```python
# Câu 4e: Hiển thị danh sách sinh viên gồm mã, họ tên, điểm trung bình và kết quả xếp loại học tập của sinh viên, 
# trong đó kết quả xếp loại học tập được xét dựa trên điểm trung bình theo quy tắc sau: 
#                                Điểm trung bình < 5 :  Yếu
#                         5 <= Điểm trung bình < 6.5 :  Trung bình
#                         6.5 <= Điểm trung bình < 8 :  Khá
#                               Điểm trung bình >= 8 :  Giỏi
conn = connect(dsn)
stmt = '''
SELECT s."Ma_sinh_vien", s."Ho"||' '||s."Ten" AS "Ho_ten", AVG(k."Diem") AS "Diem_trung_binh",
  CASE 
	WHEN AVG(k."Diem")<5 THEN 'Yếu'
	WHEN AVG(k."Diem")<6.5 THEN 'Trung bình'
	WHEN AVG(k."Diem")<8 THEN 'Khá'
	ELSE 'Giỏi'
  END AS "Xep_loai"
  FROM "SINH_VIEN" s INNER JOIN "KET_QUA" k ON s."Ma_sinh_vien"=k."Ma_sinh_vien"
GROUP BY s."Ma_sinh_vien", s."Ho"||' '||s."Ten";
'''
df = pd.read_sql(stmt, conn)
df.head()
```

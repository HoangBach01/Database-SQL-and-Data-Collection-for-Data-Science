```python
from mysql.connector import connect
```

```python
def run_sql(sql):
    cur = conn.cursor()
    cur.execute(sql)
    row = cur.fetchone()
    while row:
        print(row)
        row = cur.fetchone()    
    cur.close()
```

```python
import glob 
mydir = "EX3/"
fnames = glob.glob(mydir+'*.sql')
```

```python
conn = connect(host='localhost', port=3308, database='salesdb', user='root', password='')
```

```python
for name in fnames:
    f = open(name, 'r', encoding="UTF-8")
    script = f.read()
    f.close()
    print('*'*70)
    print('file name:', name)
    print(script)
    print('*'*70)
    run_sql(script)
```

```python
conn.close()
```
